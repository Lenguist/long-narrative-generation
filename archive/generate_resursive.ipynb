```python
def rephrase_premise(original_premise, n):
  precision = 0.2
  response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
      {"role": "system", "content": "You are an expert language model"},
      {"role": "user", "content": f"""Expand or summarize this text to be of the length {n} words. Try to adhere to length as close as possible:
      {original_premise}
      """},
    ]
  )
  new_premise = response.choices[0].message.content
  if len(new_premise.split(" ")) >= (1-precision)*n and len(new_premise.split(" ")) <= (1+precision)*n:
    print(f"new len is {len(new_premise.split(' '))}")
    return new_premise
  else:
    print("couldn't generate within precision")
    print(f"new len is {len(new_premise.split(' '))}")
    return new_premise
```

```python
n_list = [100, 50, 20, 10, 5]
all_premises = {}
for n in n_list:
  all_premises[f"len{n}"] = [rephrase_premise(premise, n) for premise in premises_original]
```

```python
# Write the dictionary to a file
with open('premises_rephrased.json', 'w') as json_file:
    json.dump(all_premises, json_file, indent=4)
```

```python
# Assuming generate_response is a function that takes a premise and a prompt
# and returns a generated story as a string. You would need to define this according to your model or API.
def generate_response(premise, prompt):
    response = client.chat.completions.create(
    model="gpt-3.5-turbo",
    messages=[
        {"role": "system", "content": "You are an expert storywriter"},
        {"role": "user", "content": f"""{prompt}:
        {premise}
        """},

    ]
    )
    print("succesfully generated")
    return response.choices[0].message.content
```

```python
import json
import pandas as pd

# List of user prompts
user_prompts = prompts[:4]

# Initialize the list that will be used to create the dataframe
data_for_df = []

# Process each premise and prompt, and store the results
for length_key, premises_list in all_premises.items():
    premise_len = int(length_key[3:])  # Extracting the numerical part of the key for length
    for premise in premises_list:
        for user_prompt in user_prompts:
            response = generate_response(premise, user_prompt)
            response_len = len(response.split())  # Count the number of words in the response
            ratio = response_len / premise_len if premise_len else 0

            # Append the data to the list for dataframe
            data_for_df.append({
                "prompt": user_prompt,
                "premise-len": premise_len,
                "premise": premise,
                "response": response,
                "response-len": response_len,
                "ratio": ratio
            })

# Convert the list to a dataframe
df = pd.DataFrame(data_for_df)

# Output the dataframe to a CSV file
df.to_csv('story_generation_output.csv', index=False)

# Also save the JSON output
with open('story_generation_output.json', 'w') as json_file:
    json.dump(data_for_df, json_file, indent=4)

print("JSON file and CSV file with generated responses have been created.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataframe from CSV just in case it's not in memory
df = pd.read_csv('story_generation_output.csv')

# Set up the matplotlib figure
sns.set(style="whitegrid")

# Distribution of response lengths
plt.figure(figsize=(10, 6))
sns.histplot(df['response-len'], bins=30, kde=False)
plt.title('Distribution of Response Lengths')
plt.xlabel('Response Length')
plt.ylabel('Frequency')
plt.show()

# Boxplot of response lengths by premise length
plt.figure(figsize=(12, 6))
sns.boxplot(x='premise-len', y='response-len', data=df)
plt.title('Response Lengths by Premise Length')
plt.xlabel('Premise Length')
plt.ylabel('Response Length')
plt.show()

# Scatter plot of premise length vs response length
plt.figure(figsize=(10, 6))
sns.scatterplot(x='premise-len', y='response-len', data=df)
plt.title('Premise Length vs Response Length')
plt.xlabel('Premise Length')
plt.ylabel('Response Length')
plt.show()

# Average ratio by premise length
avg_ratio_by_premise_len = df.groupby('premise-len')['ratio'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='premise-len', y='ratio', data=avg_ratio_by_premise_len)
plt.title('Average Ratio by Premise Length')
plt.xlabel('Premise Length')
plt.ylabel('Average Ratio (Response Len/Premise Len)')
plt.show()

# Heatmap of average response length for each prompt and premise length
# Pivot the data for heatmap
heatmap_data = df.pivot_table(values='response-len', index='prompt', columns='premise-len', aggfunc='mean')
plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", linewidths=.5)
plt.title('Heatmap of Average Response Length by Prompt and Premise Length')
plt.xlabel('Premise Length')
plt.ylabel('Prompt')
plt.show()

# Remember to show the plots
plt.show()
```
